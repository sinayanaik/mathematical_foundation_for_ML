Frobenius Norm --> Function that enables to quantify size of the Matrix.


```python
import numpy as np
X = np.array([[1,9],[3,4]])
```

```python

def fb_norm_eval(X):
    fb_norm = 0
    for i in range(0,np.shape(X)[0]):
        for j in range(0,np.shape(X)[1]):
            fb_norm += X[i][j]**2
    return np.sqrt(fb_norm)

fb_norm_eval(X)
```

```python
# using np function
np.linalg.norm(X)
```

```python
# write code to declare and evaluate l2 norm of X in pytorch and tensorflow
import torch
import tensorflow as tf

X_torch = torch.tensor([[1,9],[3,4]],dtype=torch.float32)
X_tf = tf.Variable([[1,9],[3,4]],dtype=tf.float32)

print(torch.norm(X_torch))
print(tf.norm(X_tf))
```

### Matrix Multiplication
- In Numpy : Use `numpy.dot()`


```python
A = np.array([[3,4],[5,6],[7,8]])
B = np.array([[1],[2]])

def matrix_mult(A,B):
    C = np.zeros((np.shape(A)[0],np.shape(B)[1]))
    for i in range(0,np.shape(A)[0]):
        for j in range(0,np.shape(B)[1]):
            for k in range(0,np.shape(A)[1]):
                C[i][j] += A[i][k]*B[k][j]
    return C

matrix_mult(A,B)
```

```python
np.dot(A,B)
```

In Tensorflow and Pytorch

```python
A_torch = torch.tensor([[3,4],[5,6],[7,8]],dtype=torch.float32)
B_torch = torch.tensor([[1],[2]],dtype=torch.float32)

A_tf = tf.Variable([[3,4],[5,6],[7,8]],dtype=tf.float32)
B_tf = tf.Variable([[1],[2]],dtype=tf.float32)

print(torch.matmul(A_torch,B_torch))
print(tf.matmul(A_tf,B_tf))
```
