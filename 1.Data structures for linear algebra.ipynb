Norm


```python
import numpy as np
from sympy import symbols, Eq, latex

x = np.array([1, 2, 3, 4])

# Lambda function for L1 norm
l1_norm = lambda arr: np.sum(np.abs(arr))

# Lambda function for squared L2 norm
sq_l2_norm = lambda arr: np.sum(arr**2)

print(f"$L_1$ norm : {l1_norm(x)}")
print(f"$L_2$ norm : {np.linalg.norm(x)}")
print(f"$Sq_L_2$ norm : {sq_l2_norm(x)} = {np.dot(x,x)}")

```

### Higher-Rank Tensors

As an example, rank 4 tensors are common for images, where each dimension corresponds to:

1. Number of images in training batch, e.g., 32
2. Image height in pixels, e.g., 28 for [MNIST digits](http://yann.lecun.com/exdb/mnist/)
3. Image width in pixels, e.g., 28
4. Number of color channels, e.g., 3 for full-color images (RGB)
